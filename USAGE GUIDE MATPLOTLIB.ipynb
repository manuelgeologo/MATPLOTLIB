## Figuras

La figura mantiene el seguimiento de todos los diagramas (axes), una noción de los elementos (títulos, leyenda,etc), y del lienzo (Canvas) (no tan importante para nosotros, pues solo lo usamos). Una figura puede contener un número cualquiera de diagramas (Axes), pero típicamente al menos 1. 
 
La forma más fácil de crear una nueva figura es con 'plot'

```python
fig = plt.figure() # crea una figura vacia sin diagramas
fig, ax = plt.subplots() # una figura con un solo diagrama
fig, ax = plt.subplots(2,2) # una figura con una grilla de diagramas 2x2
```

Es conveniente crear los ejes junto con la figura, pero también se puede hacer despues, permitiendo visualizaciones más complejas. 

## Axes (Diagramas)

Es lo que uno se imagina como un gŕafico 'plot'. Es la región de la imagen con el espacio de datos. Una figura dada puede tener cualquier número de diagramas, pero un diagrama puede estár solo en una Figura. Los diagramas contienen dos (o 3 en 3d) Ejes, los cuales dan los límites de los datos (esto se controla con los métodos **axes.Axes.set_xlim()** y **axes.Axes.set_ylim()**. Cada diagrama tiene un título (dado por **set_title()**), una etiqueta para cada eje (dado por **set_xlabel()**), lo mismo para eje y..

La clase Axes y sus funciones son el punto de entrada para trabajar con la interfaz OO

## Axis (ejes)

Estos son objetos tipo linea-número. se preocupan de los límites de los gráficos y generar las graduaciones (ticks) y las etiquetas de los ticks (strings). La ubicación de los graduaciones es determinada por el objeto **Locator** y los strings de las etiquetas son formateadas por un **Formatter**. La combinación de los correctores **Locator** y **Formatter** da una control muy bueno sobre la localización y etiquetas (Pruebe con uno de los gŕaficos anteriores). 

## Artist (Artista)

Básicamente todo lo que se puede ver en una figura es un 'artista' (incluso la Figura, Diagramas, y Ejes). Esto incluye objetos de **Texto**, objetos **Line2d**, objetos de **collection**, objetos de **Patch**. Cuando la figura es renderizada, todos los 'artistas' son dibujados en el lienzo (**canvas**).

## La interfaz orientada a objetos (oo) y la interfaz pyplot

Básicamente hay dos formas de usar matplotlib:

-Crear explicitamente figuras y diagramas, y llamar métodos en ellos (estilo orientado en objetos)

-Depender de **pyplot** para que cree automáticamente y maneje las figuras y diagramas, y usar las funciones pyplot para graficar. 



```python
#podriamos usar un estilo OO

x = np.linspace(0,2,100)

#notemos que incluso en el estilo OO se usa pyploy.figure para crear la figura

fig, ax = plt.subplots() # crea la figura y un diagrama
ax.plot(x,x,label='lineal') #grafica datos en el diagrama
ax.plot(x,x**2,label='cuadrática') #más datos en el diagrama
ax.plot(x,x**3,label='cubica') 
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_title('Gráfico simple') #añade título al gráfico
ax.legend() #añade una leyenda
```
